###Process image with different convolution kernels

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

def get_image(kernel):
	img=mpimg.imread('./images/shanghai.jpg')
	lum_img = img[:,:,1]
	ans = np.zeros((lum_img.shape[0] - 2, lum_img.shape[1] - 2))
	print('...')

	for i in range(1, lum_img.shape[0] - 1):
		for j in range(1, lum_img.shape[1] - 1):
			for p in range(0,3):
				for q in range(0,3):
					x = i + p - 1
					y = j + q - 1
					ans[i - 1][j - 1] = ans[i - 1][j - 1] + lum_img[x][y] * kernel[p][q]

	# I don't know whether this is reasonable
	ans[ans < ans.mean()] = ans.mean()
	return ans
```

```python
kernel1 = np.array([1, 0, -1, 0, 0, 0, -1, 0, 1]).reshape(3,3)
kernel2 = np.array([0, 1, 0, 1, -4, 1, 0, 1, 0]).reshape(3,3)
#kernel3 = np.array([-1, -1, -1, -1, 8, -1, -1, -1, -1]).reshape(3,3)
kernel3 = np.array([0,1,1,-1,0,1,-1,-1,0]).reshape(3,3)
# kernel3 = np.array([1,-2,1,1,-2,1,1,-2,1]).reshape(3,3)
```

```python
fig = plt.figure()

ax0 = fig.add_subplot(2, 2, 1)
img=mpimg.imread('./images/shanghai.jpg')
plt.imshow(img)

ax1 = fig.add_subplot(2, 2, 2)
plt.imshow(get_image(kernel1), cmap='gray')

ax2 = fig.add_subplot(2, 2, 3)
plt.imshow(get_image(kernel2), cmap='gray')

ax3 = fig.add_subplot(2, 2, 4)
plt.imshow(get_image(kernel3), cmap='gray')

plt.show()
```
